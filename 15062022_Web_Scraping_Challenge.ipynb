```python
import requests
r = requests.get('https://www.imdb.com/chart/moviemeter/?ref_=nv_mv_mpm')
print(len(r.text))
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.text)
```

```python
soup.find_all('a')
```

```python
##Most Popular Movies
```

```python
for link in soup.find_all('img'):
    print(link.get('alt'))
```
